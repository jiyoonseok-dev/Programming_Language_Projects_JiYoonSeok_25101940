```python
import random
import sys

# =====================
# í”Œë ˆì´ì–´ í´ë˜ìŠ¤
# =====================
class Player:
    def __init__(self):
        self.max_hp = 50
        self.hp = 50
        self.atk = 1
        self.enemy_hp_debuff = 0

    def is_alive(self):
        return self.hp > 0

    def heal_full(self):
        self.hp = self.max_hp

# =====================
# ê³µí†µ í•¨ìˆ˜
# =====================
def show_status(player, enemy_name, enemy_hp):
    print(f"\n[í”Œë ˆì´ì–´ HP] {player.hp}/{player.max_hp}")
    print(f"[{enemy_name} HP] {enemy_hp}\n")

def choose_part():
    parts = ["ë¨¸ë¦¬", "ëª¸í†µ", "ì™¼íŒ”", "ì˜¤ë¥¸íŒ”", "ì™¼ë‹¤ë¦¬", "ì˜¤ë¥¸ë‹¤ë¦¬"]
    print("ê³µê²© ë¶€ìœ„ë¥¼ ì„ íƒí•˜ì„¸ìš”:")
    for p in parts:
        print("-", p)
    return input(">> ").strip()

def game_over():
    print("\nğŸ’€ GAME OVER ğŸ’€")
    retry = input("ì²˜ìŒë¶€í„° ë‹¤ì‹œ ë„ì „í•˜ì‹œê² ìŠµë‹ˆê¹Œ? (y/n): ")
    if retry.lower() == 'y':
        main()
    else:
        sys.exit()

def reward_or_heal(player):
    print("\në³´ìƒì„ ì„ íƒí•˜ì„¸ìš”:   *ë²ˆí˜¸ë§Œ ì¨ì£¼ì„¸ìš” ì˜ˆ)1")
    print("1. ì²´ë ¥ ì™„ì „ íšŒë³µ")
    print("2. ê¹€ìš°ì œ êµìˆ˜ë‹˜ì˜ BIS A+ (ê³µê²©ë ¥ +5)")
    print("3. ê¶Œí˜ìœ¤ êµìˆ˜ë‹˜ì˜ ì¥í•™ê¸ˆ (ìµœëŒ€ ì²´ë ¥ +10)")
    print("4. í”„ë¡œê·¸ë˜ë° ë³„ì  í…ŒëŸ¬ (ì  ì²´ë ¥ -10)")
    choice = input(">> ")

    if choice == "1":
        player.heal_full()
        print("ì²´ë ¥ì´ ëª¨ë‘ íšŒë³µë˜ì—ˆìŠµë‹ˆë‹¤!")
    elif choice == "2":
        player.atk += 5
        print("ê³µê²©ë ¥ì´ ì¦ê°€í–ˆìŠµë‹ˆë‹¤!")
    elif choice == "3":
        player.max_hp += 10
        player.hp += 10
        print("ìµœëŒ€ ì²´ë ¥ì´ ì¦ê°€í–ˆìŠµë‹ˆë‹¤!")
    elif choice == "4":
        player.enemy_hp_debuff += 10
        print("ì•ìœ¼ë¡œ ì ë“¤ì˜ ì²´ë ¥ì´ ê°ì†Œí•©ë‹ˆë‹¤!")

print("ë‹¹ì‹ ì€ ITMì˜ íƒ‘ì— ì…ì¥í–ˆìŠµë‹ˆë‹¤.")

# =====================
# 1ì¸µ ê¹€í˜„ìš°
# =====================
def floor1(player):
    print("\n=== 1ì¸µ : ê¹€í˜„ìš° (íŠœí† ë¦¬ì–¼) ===")
    print("ITMì˜ ê¸°ê°•ì„ ì–´ì§€ë¥´ëŠ” 3ëª…ì˜ ì•…ì„ ì²˜ì¹˜í•´ì¤˜!.")

    enemy_hp = 5
    enemy_name = "ê¹€í˜„ìš°"

    while enemy_hp > 0:
        show_status(player, enemy_name, enemy_hp)
        part = choose_part()

        if part == "ë¨¸ë¦¬":
            dmg = 3
            print("ğŸ’¥ ê¹€í˜„ìš°ì˜ ì•ˆê²½ì„ ë¶€ì‰ˆë‹¤!")
        else:
            dmg = 1

        enemy_hp -= dmg
        print(f"ê¹€í˜„ìš°ì—ê²Œ {dmg} ë°ë¯¸ì§€ë¥¼ ì…í˜”ë‹¤! (ë‚¨ì€ HP: {max(enemy_hp,0)})")

        if enemy_hp <= 0:
            break

        enemy_dmg = random.randint(1, 5)
        player.hp -= enemy_dmg
        print(f"í”Œë ˆì´ì–´ê°€ {enemy_dmg} ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ë‹¤! (ë‚¨ì€ HP: {player.hp})")

        if not player.is_alive():
            game_over()

    print("\n'ì¤€ë¹„ê°€ ë˜ì—ˆêµ¬ë‚˜! ì´ê±´ ë‚˜ì˜ ì„ ë¬¼ì´ì•¼'")
    print("ğŸ ê¹€í˜„ìš°ì˜ ìì·¨ë°© ì‹ì¹¼ íšë“! (ê³µê²©ë ¥ 5)")
    player.atk = 5

# =====================
# 2ì¸µ ê³ ì§„í˜•
# =====================
def floor2(player):
    print("\n=== 2ì¸µ : ê³ ì§„í˜• ===")
    print("'ë­ì‹œëƒ?'")

    enemy_hp = max(1000 - player.enemy_hp_debuff, 1)
    enemy_name = "ê³ ì§„í˜•"

    while enemy_hp > 0:
        show_status(player, enemy_name, enemy_hp)
        part = choose_part()

        if part == "ì˜¤ë¥¸ë‹¤ë¦¬":
            print("ğŸ”¥ ê³ ì§„í˜•ì€ ì˜¤ë¥¸ìª½ ì‹­ìì¸ëŒ€ê°€ ì—†ìŠµë‹ˆë‹¤! í¬ë¦¬í‹°ì»¬ x100!")
            enemy_hp = 0
            break
        else:
            enemy_hp -= 1
            print("1 ë°ë¯¸ì§€ë¥¼ ì…í˜”ë‹¤.")

        enemy_dmg = 10
        player.hp -= enemy_dmg
        print(f"í”Œë ˆì´ì–´ê°€ {enemy_dmg} ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ë‹¤! (ë‚¨ì€ HP: {player.hp})")

        if not player.is_alive():
            game_over()

# =====================
# 3ì¸µ ìµœì¤€í¬
# =====================
def floor3(player):
    print("\n=== 3ì¸µ : ìµœì¤€í¬ ===")
    print("'SDTë¥¼ ë¬´ì‹œí•˜ì§€ë§ˆë¼.'")

    enemy_hp = max(30 - player.enemy_hp_debuff, 1)
    enemy_atk = 20
    enemy_name = "ìµœì¤€í¬"

    while enemy_hp > 0:
        show_status(player, enemy_name, enemy_hp)
        part = choose_part()

        dmg = player.atk
        if part == "ëª¸í†µ":
            dmg *= 2
            print("ğŸ’” ê·¸ì˜ ì‹¬ì¥ì€ ì—¬ë¦½ë‹ˆë‹¤..")

        enemy_hp -= dmg
        print(f"{dmg} ë°ë¯¸ì§€ë¥¼ ì…í˜”ë‹¤!")

        if enemy_hp <= 0:
            break

        if random.random() < 0.5:
            print("ğŸ˜µ ìµœì¤€í¬ê°€ ì–´ë¦¬ë²„ë¦¬í•˜ë‹¤! (ìƒëŒ€ ê³µê²©ì´ ë¹—ë‚˜ê°”ìŠµë‹ˆë‹¤)")
        else:
            print("'ê°€ì‹œê³ ~'")
            player.hp -= enemy_atk
            print(f"{enemy_atk} ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ë‹¤! (ë‚¨ì€ HP: {player.hp})")

        if not player.is_alive():
            game_over()

# =====================
# 4ì¸µ ìµœí™ì„œ
# =====================
def floor4(player):
    print("\n=== 4ì¸µ : ìµœí™ì„œ ===")
    print("'I'm spiderman'")

    enemy_hp = 1
    enemy_atk = 1
    evade = 0.7
    turn = 0

    while enemy_hp > 0:
        show_status(player, "ìµœí™ì„œ", enemy_hp)
        part = choose_part()

        if random.random() > evade:
            enemy_hp -= player.atk
            if turn == 0:
                print("'ë²Œë ˆ ì¥ ë‹´ëª¨í•œë‹¤'")
        else:
            print("ê³µê²©ì„ í”¼í–ˆë‹¤!")

        if enemy_hp <= 0:
            break

        print("'ìŠ¤íŒŒì´ë”ë§¨ ì¥ ë‹´ëª¨í•œë‹¤'")
        enemy_atk += 15
        evade = max(evade - 0.1, 0.2)

        player.hp -= enemy_atk
        print(f"{enemy_atk} ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ë‹¤! (ë‚¨ì€ HP: {player.hp})")

        if not player.is_alive():
            game_over()

        turn += 1

    print("\nğŸ‰ ë‹¹ì‹ ì€ ITMì„ ì§€ì¼°ìŠµë‹ˆë‹¤!")
    print("ğŸ™ ì—´ì‹¬íˆ í–ˆìŠµë‹ˆë‹¤.. ë³„ì  5ê°œë¥¼ ë¶€ë””..")

# =====================
# ë©”ì¸
# =====================
def main():
    player = Player()
    floor1(player)
    reward_or_heal(player)
    floor2(player)
    reward_or_heal(player)
    floor3(player)
    reward_or_heal(player)
    floor4(player)

main()
```
